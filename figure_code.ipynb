# **Figures**

```python
import os

if not os.path.exists("plots"):
    os.makedirs("plots")
```

```python
import numpy as np
```

**Calculate means and standard deviations for pLink**

```python
plink_csms_mean = np.mean(
    [
        len(plink_r1["crosslink-spectrum-matches"]),
        len(plink_r2["crosslink-spectrum-matches"]),
        len(plink_r3["crosslink-spectrum-matches"]),
    ]
)
plink_csms_sd = np.std(
    [
        len(plink_r1["crosslink-spectrum-matches"]),
        len(plink_r2["crosslink-spectrum-matches"]),
        len(plink_r3["crosslink-spectrum-matches"]),
    ]
)
plink_xls_mean = np.mean(
    [
        len(plink_r1["crosslinks"]),
        len(plink_r2["crosslinks"]),
        len(plink_r3["crosslinks"]),
    ]
)
plink_xls_sd = np.std(
    [
        len(plink_r1["crosslinks"]),
        len(plink_r2["crosslinks"]),
        len(plink_r3["crosslinks"]),
    ]
)
```

**Calculate means and standard deviations for Scout**

```python
scout_csms_mean = np.mean(
    [
        len(scout_r1["crosslink-spectrum-matches"]),
        len(scout_r2["crosslink-spectrum-matches"]),
        len(scout_r3["crosslink-spectrum-matches"]),
    ]
)
scout_csms_sd = np.std(
    [
        len(scout_r1["crosslink-spectrum-matches"]),
        len(scout_r2["crosslink-spectrum-matches"]),
        len(scout_r3["crosslink-spectrum-matches"]),
    ]
)
scout_xls_mean = np.mean(
    [
        len(scout_r1["crosslinks"]),
        len(scout_r2["crosslinks"]),
        len(scout_r3["crosslinks"]),
    ]
)
scout_xls_sd = np.std(
    [
        len(scout_r1["crosslinks"]),
        len(scout_r2["crosslinks"]),
        len(scout_r3["crosslinks"]),
    ]
)
```

**Calculate means and standard deviations for XlinkX**

```python
xlinkx_csms_mean = np.mean(
    [
        len(xlinkx_r1["crosslink-spectrum-matches"]),
        len(xlinkx_r2["crosslink-spectrum-matches"]),
        len(xlinkx_r3["crosslink-spectrum-matches"]),
    ]
)
xlinkx_csms_sd = np.std(
    [
        len(xlinkx_r1["crosslink-spectrum-matches"]),
        len(xlinkx_r2["crosslink-spectrum-matches"]),
        len(xlinkx_r3["crosslink-spectrum-matches"]),
    ]
)
xlinkx_xls_mean = np.mean(
    [
        len(xlinkx_r1["crosslinks"]),
        len(xlinkx_r2["crosslinks"]),
        len(xlinkx_r3["crosslinks"]),
    ]
)
xlinkx_xls_sd = np.std(
    [
        len(xlinkx_r1["crosslinks"]),
        len(xlinkx_r2["crosslinks"]),
        len(xlinkx_r3["crosslinks"]),
    ]
)
```

## **Figure 2A**

*****

```python
import matplotlib.pyplot as plt

se = ("pLink", "Scout", "XlinkX")
se_means = {
    "CSMs": (plink_csms_mean, scout_csms_mean, xlinkx_csms_mean),
    "XLs": (plink_xls_mean, scout_xls_mean, xlinkx_xls_mean),
}
se_sds = {
    "CSMs": (plink_csms_sd, scout_csms_sd, xlinkx_csms_sd),
    "XLs": (plink_xls_sd, scout_xls_sd, xlinkx_xls_sd),
}

x = np.arange(len(se))
width = 0.4
multiplier = 0

fig, ax = plt.subplots(layout="constrained", figsize=(7, 4))

colors = ["#6d4bff", "#ac99ff"]
for attribute, measurement in se_means.items():
    offset = width * multiplier
    rects = ax.bar(
        x + offset, measurement, width, label=attribute, color=colors[multiplier % 2]
    )
    err = ax.errorbar(
        x + offset, measurement, yerr=se_sds[attribute], fmt="o", color="black"
    )
    ax.bar_label(rects, padding=14, fmt="{:.0f}")
    multiplier += 1

ax.set_xlabel("Crosslink Search Engine")
ax.set_ylabel("Average Number of CSMs and XLs (n=3)")
ax.set_title("Crosslink-Spectrum-Matches (CSMs) and Crosslinks (XLs) by Search Engine")
ax.set_xticks(x + width / 2, se)
ax.legend(loc="upper right", ncols=1)
ax.set_ylim(0, 860)

plt.savefig(
    "plots/figure2a.svg",
    dpi=300,
    transparent=True,
    bbox_inches="tight",
)
```

```python
plink_intersection = pyXLMS.transform.intersection(
    pyXLMS.transform.intersection(
        plink_r1["crosslinks"], plink_r2["crosslinks"], by="protein", use="data_a"
    ),
    plink_r3["crosslinks"],
    by="protein",
    use="data_a",
)
```

```python
scout_intersection = pyXLMS.transform.intersection(
    pyXLMS.transform.intersection(
        scout_r1["crosslinks"], scout_r2["crosslinks"], by="protein"
    ),
    scout_r3["crosslinks"],
    by="protein",
)
```

## **Export to XMAS**

*****

**Filter for CAS9 intra-crosslinks**

```python
plink_intersection_cas9 = pyXLMS.transform.filter_proteins(
    plink_intersection, proteins=["CAS9HALO"]
)["Both"]
scout_intersection_cas9 = pyXLMS.transform.filter_proteins(
    scout_intersection, proteins=["CAS9HALO"]
)["Both"]
xlinkx_intersection_cas9 = pyXLMS.transform.filter_proteins(
    xlinkx_intersection, proteins=["CAS9HALO"]
)["Both"]
```

**Export to XMAS**

```python
pyXLMS.exporter.to_xmas(plink_intersection_cas9, filename="export/xmas_plink");
```

```python
pyXLMS.exporter.to_xmas(scout_intersection_cas9, filename="export/xmas_scout");
```

```python
pyXLMS.exporter.to_xmas(xlinkx_intersection_cas9, filename="export/xmas_xlinkx");
```

## **Export to XMAS & AlphaLink2**

*****

**Create intersection of all CAS9 intra-crosslinks found in all three replicates of all search engine results**

```python
all_intersection = pyXLMS.transform.intersection(
    pyXLMS.transform.intersection(
        plink_intersection_cas9, scout_intersection_cas9, by="protein", use="data_a"
    ),
    xlinkx_intersection_cas9,
    by="protein",
    use="data_a",
)
```

**Export to XMAS**

```python
pyXLMS.exporter.to_xmas(all_intersection, filename="export/xmas_all");
```

**Export to AlphaLink2**

```python
pyXLMS.exporter.to_alphalink2(
    all_intersection, fasta="fasta/cas9.fasta", filename_prefix="export/al2_all"
);
```
